```python
#reading stocks csv files
from functools import reduce
delta = "DAL.csv"
hilton = "HLT.csv"
singapore = 'SINGF.csv'
marriott = "VAC.csv"
delta_df = pd.read_csv(delta, usecols = ['Date','Adj Close'])
delta_df = delta_df.rename(columns={'Adj Close': 'Close_DELTA'})
hilton_df = pd.read_csv(hilton,usecols = ['Date','Adj Close'])
hilton_df = hilton_df.rename(columns={'Adj Close': 'Close_HILT'})
singapore_df = pd.read_csv(singapore,usecols = ['Date','Adj Close'])
singapore_df = singapore_df.rename(columns={'Adj Close': 'Close_SING'})
marriott_df = pd.read_csv(marriott,usecols = ['Date','Adj Close'])
marriott_df = marriott_df.rename(columns={'Adj Close': 'Close_MRT'})
data_frames = [delta_df, hilton_df, marriott_df,singapore_df]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],how='outer'), data_frames)
df_merged
```

```python
#cleaning the stock dataframe
df_merged['Date'] = pd.to_datetime(df_merged['Date'])
#df_merged['Date'] = df_merged['Date'].dt.date
df_merged= df_merged.sort_values(by='Date')
df_merged.reset_index()
df_merged.dropna()
df_merged
```

```python
df_merged.drop(df_merged.head(234).index, inplace=True)
df_merged=df_merged.reset_index()
df_merged=df_merged.drop(columns=['index'])
df_merged.dropna()
df_merged
```
